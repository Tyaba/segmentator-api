```python
from pathlib import Path

from PIL import Image
from ultralytics import YOLOWorld

from segmenter_api.utils.image import draw_bboxes
```

```python
model_path = Path("models/yolov8x-worldv2.pt")
```

```python
# Load a pretrained YOLO11n model
model = YOLOWorld(model_path)
```

```python
image_path = Path("data/multiple_products.png")
image = Image.open(image_path)
```

```python
# Perform object detection on an image
results = model(image_path)  # Predict on an image
```

```python
results[0].save("yolo.png")
```

```python
draw_bboxes(image, results[0].boxes.xyxy.tolist())
```
